### Define analysis functions:

- ***freq_per_table(df, column_name)***: frequency and percentage function
- **Cross-tabulation --> 2 variables**:
  - ***crosstabulate(df, row_var, col_var)***: cross-tabulation with frequencies and percentages
  - ***crosstab_summary(df, row_var, col_var, output="frequency")***: cross-tabulation with frequencies only
  - ***crosstab_summary(df, row_var, col_var, output="percentage")***: cross-tabulation with percentages only
- **Cross-tabulation --> 1 vs multiple variables**:  
  - ***crosstab_multiple(df, row_var, col_vars, output="frequency")***
  - ***crosstab_multiple(df, row_var, col_vars, output="percentage")***
- **Missing data analysis**:
  - ***missing_data(df, variables)***
  - ***missing_data_group(df, group_var, variables)***
- **Summary stats**:
  - ***summary_stats(df, numeric_only=True)***
  - ***summary_stats_for_vars(df, vars_list, numeric_only=True)***



```python
def freq_per_table(df, column_name):
    """
    Generate a frequency and proportion table for a given column in a DataFrame, including totals.

    Parameters:
    - df (pd.DataFrame): The DataFrame containing the data.
    - column_name (str): The name of the column for which to compute the table.

    Returns:
    - pd.DataFrame: A DataFrame with 'Frequency' and 'Proportion' columns, including totals.
    """
    frequency = df[column_name].value_counts(dropna=False)  # Include missing values
    percentage = 100*df[column_name].value_counts(normalize=True, dropna=False)

    # Create the result DataFrame
    result = pd.DataFrame({'Frequency': frequency, 'Percentage': percentage})

    # Add a Total row
    total_row = pd.DataFrame({
        'Frequency': [frequency.sum()],
        'Percentage': [percentage.sum()]
    }, index=['Total'])

    # Append the total row to the result
    result = pd.concat([result, total_row])

    return result

# Example usage
result = freq_per_table(df, 'tenure')
result['Proportion'] = result['Proportion'] * 100
print(result)
```

```python
import pandas as pd

def crosstabulate(df, row_var, col_var):
    """
    Generate a cross-tabulation table with frequencies and proportions for two variables.

    Parameters:
    - df (pd.DataFrame): The DataFrame containing the data.
    - row_var (str): The column name to be used as the row variable.
    - col_var (str): The column name to be used as the column variable.

    Returns:
    - pd.DataFrame: A DataFrame showing frequencies and proportions.
    """
    # Cross-tabulate with frequencies
    freq_table = pd.crosstab(df[row_var], df[col_var], dropna=False)

    # Add totals for rows and columns
    freq_table['Row Total'] = freq_table.sum(axis=1)
    freq_table.loc['Column Total'] = freq_table.sum(axis=0)

    # Proportions (normalize over the entire table)
    prop_table = pd.crosstab(df[row_var], df[col_var], normalize=True, dropna=False) * 100
    prop_table['Row Total'] = prop_table.sum(axis=1)
    prop_table.loc['Column Total'] = prop_table.sum(axis=0)

    # Combine frequency and proportion tables into a single table
    combined_table = freq_table.astype(str) + " (" + prop_table.round(2).astype(str) + "%)"

    return combined_table

# Example usage
result = crosstabulate(df, 'treatment_embed', 'treatment_track')
print(result)
```

```python
import pandas as pd

def crosstab_summary(df, row_var, col_var, output="frequency"):
    """
    Generate a cross-tabulation table with either frequencies or percentages.

    Parameters:
    - df (pd.DataFrame): The DataFrame containing the data.
    - row_var (str): The column name to be used as the row variable.
    - col_var (str): The column name to be used as the column variable.
    - output (str): Specify "frequency" for raw counts or "percentage" for proportions.

    Returns:
    - pd.DataFrame: A DataFrame showing either frequencies or percentages.
    """
    if output == "frequency":
        # Cross-tabulate with raw frequencies
        table = pd.crosstab(df[row_var], df[col_var], dropna=False)
    elif output == "percentage":
        # Cross-tabulate with proportions (normalized over the entire table)
        table = round(pd.crosstab(df[row_var], df[col_var], normalize=True, dropna=False) * 100, 2)
    else:
        raise ValueError("Invalid output type. Use 'frequency' or 'percentage'.")

    # Add totals for rows and columns
    table['Row Total'] = table.sum(axis=1)
    table.loc['Column Total'] = table.sum(axis=0)

    return table

# Example usage
freq_result = crosstab_summary(df, 'treatment_embed', 'treatment_track', output="frequency")
print("Frequency Table:")
print(freq_result)

percentage_result = crosstab_summary(df, 'treatment_embed', 'treatment_track', output="percentage")
print("\nPercentage Table:")
print(percentage_result)
```

```python
def crosstab_multiple(df, row_var, col_vars, output="frequency"):
    """
    Cross-tabulate one variable against multiple variables.

    Parameters:
    - df (pd.DataFrame): The DataFrame containing the data.
    - row_var (str): The column name to use as the row variable.
    - col_vars (list of str): A list of column names to use as the column variables.
    - output (str): Specify "frequency" for raw counts or "percentage" for proportions.

    Returns:
    - dict: A dictionary of DataFrames, each representing a cross-tabulation.
    """
    result = {}

    for col_var in col_vars:
        if output == "frequency":
            # Frequency cross-tabulation
            table = pd.crosstab(df[row_var], df[col_var], dropna=False)
        elif output == "percentage":
            # Percentage cross-tabulation
            table = pd.crosstab(df[row_var], df[col_var], normalize=True, dropna=False) * 100
        else:
            raise ValueError("Invalid output type. Use 'frequency' or 'percentage'.")

        # Add totals
        table['Row Total'] = table.sum(axis=1)
        table.loc['Column Total'] = table.sum(axis=0)

        # Store the table in the result dictionary
        result[col_var] = table

    return result

# Example usage
col_vars = ['treatment_track', 'facilitator_track']  # List of column variables
result = crosstab_multiple(df, 'treatment_embed', col_vars, output="frequency")

# Print the cross-tabulation results
for col_var, table in result.items():
    print(f"Cross-tabulation with column variable: {col_var}")
    print(table, "\n")
```

```python
def summary_stats(df, columns):
    """
    Generate summary statistics (mean, p25, p50, p75, min, max, N) for multiple columns in a DataFrame.

    Parameters:
    - df (pd.DataFrame): The DataFrame containing the data.
    - columns (list): A list of column names to summarize.

    Returns:
    - pd.DataFrame: A DataFrame with summary statistics for each column.
    """
    stats = {}

    for column in columns:
        stats[column] = {
            'Mean': round(df[column].mean(), 3),
            'Median': round(df[column].median(), 3),
            'P25': round(df[column].quantile(0.25), 3),
            'P75': round(df[column].quantile(0.75), 0),
            'Min': df[column].min(),
            'Max': df[column].max(),
            'N': df[column].count()
        }

    # Convert stats dictionary into a DataFrame
    result = pd.DataFrame(stats).T  # Transpose to have variables as rows
    return result

result = summary_stats(df, ['welfare_avg', 'meritocracy_avg', 'paternalism_avg'])
print(result)
```

```python
def missing_data(df, variables):
    """
    Create a table summarizing the number and percentage of missing observations for selected variables.

    Parameters:
    - df (pd.DataFrame): The DataFrame containing the data.
    - variables (list of str): A list of column names to include in the summary.

    Returns:
    - pd.DataFrame: A summary table with the number and percentage of missing observations for each variable.
    """
    # Calculate the number of missing values for each variable
    missing_counts = df[variables].isnull().sum()

    # Calculate the percentage of missing values for each variable
    total_rows = len(df)
    missing_percentages = (missing_counts / total_rows) * 100

    # Create a summary DataFrame
    summary_df = pd.DataFrame({
        'Variable': variables,
        'Missing Count': missing_counts.values,
        'Missing %': missing_percentages.values
    })

    # Sort the table by the number of missing values (optional)
    summary_df = summary_df.sort_values(by='Missing Count', ascending=False).reset_index(drop=True)

    return summary_df

# Example usage
variables_to_check = ['treatment_embed', 'treatment_track', 'facilitator_track']
missing_summary = missing_data(df, variables_to_check)
print(missing_summary)
```

```python
def missing_data_group(df, group_var, variables):
    """
    Create a table summarizing the number and percentage of missing observations for selected variables by group.

    Parameters:
    - df (pd.DataFrame): The DataFrame containing the data.
    - group_var (str): The column name for the grouping variable (e.g., treatment status).
    - variables (list of str): A list of column names to include in the summary.

    Returns:
    - pd.DataFrame: A summary table with the number and percentage of missing observations for each variable, grouped by the specified group_var.
    """
    # Group the DataFrame by the specified grouping variable (e.g., treatment status)
    grouped = df.groupby(group_var)

    # Initialize an empty list to store the results
    summary_data = []

    # Loop over the groups
    for group_name, group_data in grouped:
        # For each variable, calculate missing counts and percentages within the group
        missing_counts = group_data[variables].isnull().sum()
        total_rows = len(group_data)
        missing_percentages = (missing_counts / total_rows) * 100

        # Append the summary for the current group
        for var in variables:
            summary_data.append({
                'Group': group_name,
                'Variable': var,
                'Missing Count': missing_counts[var],
                'Missing %': missing_percentages[var]
            })

    # Convert the list of dictionaries to a DataFrame
    summary_df = pd.DataFrame(summary_data)

    return summary_df

# Example usage
group_var = 'treatment_dum'  # Specify the grouping variable (e.g., treatment status)
variables_to_check = ['treatment_embed', 'facilitator_track', 'treatment_track']
missing_summary_by_treatment = missing_data_group(df, group_var, variables_to_check)
print(missing_summary_by_treatment)
```

```python
def summary_stats(df, numeric_only=True):
    """
    Generate summary statistics for a pandas DataFrame.

    Parameters:
        df (pd.DataFrame): The DataFrame for which to calculate statistics.
        numeric_only (bool): Whether to include only numeric columns. Default is True.

    Returns:
        pd.DataFrame: A summary statistics table.
    """
    if numeric_only:
        df = df.select_dtypes(include=['number'])  # Filter numeric columns

    # Generate summary statistics
    summary = df.describe().T  # Transpose to make rows the column names
    summary['missing_values'] = df.isnull().sum()  # Add missing value count
    summary['unique_values'] = df.nunique()  # Add unique value count

    return summary

# print(summary_stats(df))
```

```python
def summary_stats_for_vars(df, vars_list, numeric_only=True):
    """
    Generate summary statistics for a specified list of variables in a DataFrame.

    Parameters:
        df (pd.DataFrame): The DataFrame for which to calculate statistics.
        vars_list (list): List of column names to include in the summary.
        numeric_only (bool): Whether to include only numeric columns. Default is True.

    Returns:
        pd.DataFrame: A summary statistics table for the selected variables.
    """
    # Select only the columns in the vars_list
    selected_df = df[vars_list]

    if numeric_only:
        selected_df = selected_df.select_dtypes(include=['number'])  # Filter numeric columns

    # Generate summary statistics
    summary = selected_df.describe().T  # Transpose to make rows the column names
    summary['missing_values'] = selected_df.isnull().sum()  # Add missing value count
    summary['unique_values'] = selected_df.nunique()  # Add unique value count

    return summary

list_sumstats = ['meritocracy_avg', 'welfare_avg']
print(summary_stats_for_vars(df, list_sumstats))
```

```python
import scipy.stats as stats

def summary_stats_by_group(df, group_col, summary_cols):
    """
    Generate summary statistics (including confidence intervals) for specified columns, grouped by a group column, in long format.

    Parameters:
        df (pd.DataFrame): The DataFrame to calculate summary statistics on.
        group_col (str): The column to group by.
        summary_cols (list): List of columns for which to calculate summary statistics.

    Returns:
        pd.DataFrame: A long-format DataFrame with summary statistics for each group and variable.
    """
    # Initialize an empty list to store results
    results = []

    # Loop through each variable and compute statistics
    for col in summary_cols:
        grouped = df.groupby(group_col)[col].agg(['mean', 'median', 'std', 'count', 'min', 'max']).reset_index()

        # Calculate confidence intervals for the mean (95%)
        grouped['ci_lower'] = grouped.apply(
            lambda x: x['mean'] - stats.t.ppf(0.975, df=int(x['count']) - 1) * (x['std'] / (x['count'] ** 0.5))
            if x['count'] > 1 else None, axis=1
        )
        grouped['ci_upper'] = grouped.apply(
            lambda x: x['mean'] + stats.t.ppf(0.975, df=int(x['count']) - 1) * (x['std'] / (x['count'] ** 0.5))
            if x['count'] > 1 else None, axis=1
        )

        # Add a column for the variable name and reorder columns
        grouped['variable'] = col
        grouped = grouped[[
            'variable', group_col, 'mean', 'ci_lower', 'ci_upper', 'median', 'std', 'min', 'max', 'count'
        ]]
        results.append(grouped)

    # Combine all results into a single DataFrame
    summary_group = pd.concat(results, ignore_index=True)
    return summary_group

list_sumstats = ['meritocracy_avg', 'welfare_avg']
summary_stats_by_group(df, 'treatment_dum', list_sumstats)
```
